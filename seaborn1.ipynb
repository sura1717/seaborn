```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
#now plot a scatter plot
sns.scatterplot(data=tips,x="total_bill",y="tip")
```

```python
# accordin to sex divided and scatter style according to time
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex",style="time",size="size")
```

```python
#relplot --> figure level --> square shape
sns.relplot(data=tips,x="total_bill",y="tip",kind="scatter",hue="sex")
```

**line plot**

```python
sns.relplot(data=tips,x="total_bill",y="tip",kind="line",hue="sex")
```

```python
gap=px.data.gapminder()
gap.head()
```

```python
india = gap[gap["country"] == "India"]
india.head()
```

```python
sns.lineplot(data=india,x="year",y="lifeExp")
```

```python
sns.relplot(data=india,x="year",y="lifeExp",kind="line")
```

```python
tempdf=gap[gap["country"].isin(["India","Brazil","Afghanistan"])]
tempdf.head()
```

```python
#plot multiple line in a single graph
sns.lineplot(data=tempdf,x="year",y="lifeExp",hue="country",style="continent")
```

```python
sns.relplot(kind="line",data=tempdf,x="year",y="lifeExp",hue="country")
```

```python
#facet plot --> figure level function --> work with replot
#it will not work in scatterplot and line plot
sns.relplot(data=tips,x="total_bill",y="tip",col="sex",row="day",kind="line")
```

```python
#column wrap
sns.relplot(kind="scatter",data=gap,x="lifeExp",y="gdpPercap",col="year")
```

```python
sns.relplot(kind="scatter",data=gap,x="lifeExp",y="gdpPercap",col="year",col_wrap=3)
```

#**Distribution plot**

. used for univarient analysis         
. used to find out the distribution      
. Range of the observation     
. centeral tendency     
. is the data bimodal     
. Are there outliner     

**plot under distribution plot**
.histplot    
.kdeplot     
.rugplot     

```python
#figure level --> displot
# axis level --> histplot,kdeplot,rugplot
```

```python
#ploting univerient histogram
sns.histplot(data=tips,x="total_bill")
```

```python
# using figure level
sns.displot(data=tips,x="total_bill",kind="hist",bins=4,color="green")
```

.>>  its also possible to visulize the distribution of a categorical variable using the logic of a histogram      
.>>  discrete bins are automatically set for categorical variable      


```python
#countplot
sns.displot(data=tips,x="day",kind="hist")
```

```python
#hue parameter
sns.displot(data=tips,kind="hist",x="tip",hue="sex")
```

```python
sns.displot(data=tips,kind="hist",x="tip",hue="sex",element="step")
```

```python
titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
sns.displot(data=titanic,x="age",kind="hist",hue="sex")
```

```python
sns.displot(data=titanic,x="age",kind="hist",hue="sex",element="step")
```

```python
#plot diferent graph for different sex
sns.displot(data=titanic,x="age",kind="hist",hue="sex",col="sex")
```

```python
#kde plot
# rather then using discrete bin, a kde plot smooths the obsorbation with a gaussian kernal, producing a continious distinct estimate
sns.kdeplot(data=tips,x="total_bill")
```

```python
sns.kdeplot(data=tips,x="total_bill",y="tip")
```

```python
from html import escape
sns.displot(data=tips,x="total_bill",kind="kde",hue="sex",fill=True,aspect=2,height=10)
```

```python
#rugplot
# plot marginal distribution bydrawing ticks along the x and y axis
# the function is intended to complement other plots by showing the location of indivisual obserbation on an universive way

sns.displot(data=tips,x="tip",kind="kde",fill=True)
sns.rugplot(data=tips,x="tip",color="green")
#sns.displot(data=tips,x="tip",kind="rug")     this is not a figure levelo graph
```

**2D matrix plot**         
1. heatmap      
2. clusterplot

.>>  plot a rectangulor data as a color encoded matrix

```python
temp_df=gap.pivot(index="country",columns="year",values="lifeExp")
```

```python
#plot a heatmap
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# draw all uripion contry heatmap
europe=gap[gap["continent"]=="Europe"].pivot(index="country",columns="year",values="lifeExp")

plt.figure(figsize=(15,20))
sns.heatmap(europe,linewidth=0.6,linecolor="black",cmap="summer")
```

```python
# annot function parameter
plt.figure(figsize=(15,20))
sns.heatmap(europe,linewidth=0.6,linecolor="black",annot=True,cmap="summer")
```
